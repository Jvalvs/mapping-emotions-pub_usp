```python
from google.colab import drive
import sys
drive.mount('/content/gdrive')

from google_play_scraper import Sort, reviews_all
from os import listdir
from os.path import isfile, join
import pickle
from tqdm.notebook import tqdm
import pandas as pd
```

```python
myPath ='/content/gdrive/MyDrive/teste3'
sys.path.insert(1, myPath)

result = reviews_all(
    'com.c6bank.app',
    myPath,
    sleep_milliseconds=0, # defaults to 0
    lang='pt', # defaults to 'en'
    country='br', # defaults to 'us'
    sort=Sort.MOST_RELEVANT, # defaults to Sort.MOST_RELEVANT
    #filter_score_with=5 # defaults to None(means all score)
)

onlyfiles = [f for f in listdir(myPath) if isfile(join(myPath, f))]

data = []
for f in onlyfiles:
  d = pickle.load(open( myPath+ '/' +f, "rb" ))
  data += d

df = pd.DataFrame(data)
# Gera o CSV com as informa√ß√µes extra√≠das
df.to_csv('/content/gdrive/MyDrive/teste3/C6.csv')
df
```
